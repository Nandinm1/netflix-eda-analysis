# Project 01: Exploratory Data Analysis & Insights Report  
## Dataset: Netflix Movies & TV Shows Dataset

**Dataset Source:** Kaggle — Netflix Movies & TV Shows Dataset  
**Goal:** Analyze Netflix content trends, clean the dataset, create visualizations, and write data-backed insights.

## Step 1 — Load & Inspect the Data

Upload `netflix_titles.csv` to Colab before running this notebook.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

df = pd.read_csv("netflix_titles.csv")
df.head()
```

```python
print("Shape:", df.shape)
print("\nData Types:")
print(df.dtypes)
print("\nMissing Values:")
print(df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

### 5-Line Data Summary

1. The dataset contains Netflix titles including Movies and TV Shows.
2. It includes details such as title, director, cast, country, date added, release year, rating, duration, listed genres, and description.
3. Each row represents one Netflix title.
4. Some fields such as director, cast, country, rating, and date added contain missing values.
5. The dataset can be used to study Netflix content type, release trends, country distribution, ratings, and genre patterns.

## Step 2 — Clean the Data

Cleaning decisions:
- Filled missing `director`, `cast`, and `country` values with `"Unknown"` because these are descriptive categorical fields.
- Filled missing `rating` with the most frequent rating because rating is important for analysis.
- Dropped rows with missing `date_added` because date-based trend analysis requires valid dates.
- Converted `date_added` to datetime format.
- Created `year_added` and `month_added` columns for time-based analysis.
- Removed duplicate rows if any exist.

```python
df_clean = df.copy()

df_clean["director"] = df_clean["director"].fillna("Unknown")
df_clean["cast"] = df_clean["cast"].fillna("Unknown")
df_clean["country"] = df_clean["country"].fillna("Unknown")

df_clean["rating"] = df_clean["rating"].fillna(df_clean["rating"].mode()[0])

df_clean = df_clean.dropna(subset=["date_added"])

df_clean["date_added"] = pd.to_datetime(df_clean["date_added"].str.strip(), errors="coerce")
df_clean = df_clean.dropna(subset=["date_added"])

df_clean["year_added"] = df_clean["date_added"].dt.year
df_clean["month_added"] = df_clean["date_added"].dt.month

df_clean = df_clean.drop_duplicates()

print("Cleaned Shape:", df_clean.shape)
print("\nMissing Values After Cleaning:")
print(df_clean.isnull().sum())
```

## Step 3 — Exploratory Data Analysis

### Question 1: How many Movies and TV Shows are available on Netflix?

```python
type_counts = df_clean["type"].value_counts()
type_counts
```

### Question 2: Which countries produce the most Netflix content?

```python
top_countries = df_clean["country"].value_counts().head(10)
top_countries
```

### Question 3: Which ratings are most common?

```python
rating_counts = df_clean["rating"].value_counts()
rating_counts
```

### Question 4: How many titles were added each year?

```python
titles_by_year = df_clean.groupby("year_added")["show_id"].count()
titles_by_year
```

### Question 5: What are the most common genres/categories?

```python
genres = df_clean["listed_in"].str.split(", ").explode()
top_genres = genres.value_counts().head(10)
top_genres
```

## Step 4 — Visualizations

### Chart 1: Bar Chart — Movies vs TV Shows

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df_clean, x="type")
plt.title("Count of Movies vs TV Shows on Netflix")
plt.xlabel("Content Type")
plt.ylabel("Number of Titles")
plt.show()
```

### Chart 2: Line Chart — Titles Added Over Time

```python
plt.figure(figsize=(10,6))
titles_by_year.plot(kind="line", marker="o")
plt.title("Netflix Titles Added Over Time")
plt.xlabel("Year Added")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

### Chart 3: Histogram — Release Year Distribution

```python
plt.figure(figsize=(10,6))
plt.hist(df_clean["release_year"], bins=30)
plt.title("Distribution of Netflix Title Release Years")
plt.xlabel("Release Year")
plt.ylabel("Number of Titles")
plt.show()
```

### Chart 4: Scatter Plot — Release Year vs Year Added

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df_clean, x="release_year", y="year_added", hue="type", alpha=0.6)
plt.title("Release Year vs Year Added on Netflix")
plt.xlabel("Release Year")
plt.ylabel("Year Added")
plt.legend(title="Type")
plt.show()
```

### Chart 5: Pie Chart — Content Type Share

```python
plt.figure(figsize=(7,7))
type_counts.plot(kind="pie", autopct="%1.1f%%", startangle=90)
plt.title("Share of Movies and TV Shows on Netflix")
plt.ylabel("")
plt.show()
```

### Chart 6: Heatmap — Monthly Content Additions by Year

```python
monthly_pivot = df_clean.pivot_table(
    index="year_added",
    columns="month_added",
    values="show_id",
    aggfunc="count",
    fill_value=0
)

plt.figure(figsize=(12,7))
sns.heatmap(monthly_pivot, cmap="Blues", annot=False)
plt.title("Netflix Content Additions by Year and Month")
plt.xlabel("Month Added")
plt.ylabel("Year Added")
plt.show()
```

### Extra Chart 7: Top 10 Genres

```python
plt.figure(figsize=(12,6))
top_genres.sort_values().plot(kind="barh")
plt.title("Top 10 Netflix Genres")
plt.xlabel("Number of Titles")
plt.ylabel("Genre")
plt.show()
```

## Step 5 — Insights Report

1. **Movies dominate Netflix content** based on the bar chart and pie chart comparing Movies vs TV Shows. This shows Netflix has historically focused more on movie availability than TV series.
2. **Content additions increased strongly in recent years** based on the line chart of titles added by year. This suggests Netflix expanded its catalog rapidly during peak streaming growth years.
3. **The United States appears as the top content-producing country** based on the top-country analysis. This shows Netflix’s catalog is strongly influenced by U.S. content.
4. **Recent release years are more common** based on the release year histogram. This means Netflix’s catalog has a strong focus on modern and recently produced content.
5. **International Movies and Dramas are among the most common categories** based on the genre analysis. This suggests Netflix serves a global audience and invests in broad entertainment categories.

## Most Surprising Finding

The most surprising finding was how strongly Netflix content additions increased during recent years. The line chart shows that Netflix rapidly expanded its catalog instead of adding content at a steady rate. This indicates a major growth strategy focused on increasing streaming content availability.
